```python
# load data
from sklearn import datasets
boston = datasets.load_boston()
print(boston.keys())
```

```python
# summary of data
feature = boston.data
price = boston.target
print('data size = ', feature.shape)
print('target size = ', price.shape)
print('feature attributes: ', boston.feature_names)
print(boston.DESCR)
```

```python
# more details of data
import pandas as pd
df_feature = pd.DataFrame(feature, columns = boston.feature_names)
df_target = pd.DataFrame(price, columns =['MEDV'])
df_boston = pd.concat([df_feature, df_target,], axis = 1)
```

```python
# train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, price, test_size=0.3, random_state=8)
```

```python
# 2.3 linear regression and ridge regression
import numpy as np

def least_square(X, y):
    #TODO
    #closed form solution theta = (<X>X)^1(<X>Y)
    Xt = np.transpose(X)
    theta = np.matmul( np.linalg.inv(np.matmul(Xt,X)), np.matmul(Xt,y)   )
    return theta

def ridge_reg(X, y, eta):
    #TODO
    #closed form solution theta = (<X>X + eta.I)^1(<X>Y)
    Xt = np.transpose(X)
    I = np.identity(Xt.shape[0])      
    theta = np.matmul( (np.linalg.inv(np.matmul(Xt,X) + eta*I ) ), np.matmul(Xt,y) )

    return theta

# apply linear regression
theta = least_square(X_train, y_train)
df_theta = pd.DataFrame(list(zip(boston.feature_names, theta)),columns=['Feature','Coeff'])
print("The different coefficients for Linear Regression are - \n", df_theta)


eta = [20]
# apply ridge regression
for e in eta:
    theta_r = ridge_reg(X_train, y_train, e)
    df_theta_r = pd.DataFrame(list(zip(boston.feature_names, theta_r)),columns=['Feature','Coeff'])
    print("The coefficients for Ridge Regression for eta = ",e , "- \n", df_theta_r)
```

```python
# 2.4 evaluation
def pred_fn(X, theta):
    #TODO
    pred = np.matmul(X, theta) 
    
    return pred

def root_mean_square_error(pred, y):
    #TODO
    n = pred.shape[0]
    rmse = np.sqrt( (1/n)*np.sum (np.square(np.subtract(y,pred)) )  )
    return rmse

p = pred_fn(X_train,theta)
rmse = root_mean_square_error(p, y_train)

#More to do, look at 

#Least Square
theta_train = least_square(X_train, y_train)
p_train = pred_fn(X_train, theta_train)
rmse_train = root_mean_square_error(p_train, y_train)

p_test = pred_fn(X_test, theta_train)
rmse_test = root_mean_square_error(p_test, y_test)

print("RMSE Errors for Linear Regression: Train, Test - ", rmse_train, rmse_test)


# Ridge Regression
theta_train = ridge_reg(X_train, y_train, 20.0)
p_train = pred_fn(X_train, theta_train)
rmse_train_r = root_mean_square_error(p_train, y_train)

p_test = pred_fn(X_test, theta_train)
rmse_test_r = root_mean_square_error(p_test, y_test)

print("RMSE Errors for Ridge Regression : Train, Test - ", rmse_train_r, rmse_test_r)


```

```python
# 2.5 linear models of top-3 features
# linear regression using top-3 features
    #TODO

X_train_top3 = X_train[:,[5,10,12]]

print(X_train_top3.shape)
X_test_top3 = X_test[:,[5,10,12]]

theta_train_top3 = least_square(X_train_top3, y_train)

p_train_top3 = pred_fn(X_train_top3, theta_train_top3)
rmse_train_top3 = root_mean_square_error(p_train_top3, y_train)


p_test_top3 = pred_fn(X_test_top3, theta_train_top3)
rmse_test_top3 = root_mean_square_error(p_test_top3, y_test)

print("RMSE Errors for Linear Regression using Top3: Train, Test - ", rmse_train_top3, rmse_test_top3)


# ridge regression using top-3 features
    #TODO
theta_r_train_top3 = ridge_reg(X_train_top3, y_train,eta)

pr_train_top3 = pred_fn(X_train_top3, theta_r_train_top3)
rmseR_train_top3 = root_mean_square_error(pr_train_top3, y_train)

pr_test_top3 = pred_fn(X_test_top3, theta_r_train_top3)
rmseR_test_top3 = root_mean_square_error(pr_test_top3, y_test)

print("RMSE Errors for Ridge Regression using Top3: Train, Test - ", rmseR_train_top3, rmseR_test_top3)
```
